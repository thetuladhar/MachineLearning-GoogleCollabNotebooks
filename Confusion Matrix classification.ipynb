```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
red=np.random.uniform(4,10,100)
blue=np.random.uniform(0,6,100)
```

```python
t=5
plt.scatter(red,np.ones(100),color="red")
plt.scatter(blue,np.zeros(100),color="blue")
plt.plot([t,t],[0,1],color="black")
plt.plot([0,10],[.5,.5],color="black")
plt.show()
```

```python
TP=list(filter(lambda x:x>t,red))
len(TP)
```

```python
FP=list(filter(lambda x:x<=t,red))
len(FP)
```

```python
TN=list(filter(lambda x:x<=t,blue))
len(TN)
```
