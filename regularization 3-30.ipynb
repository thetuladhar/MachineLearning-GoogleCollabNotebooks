```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
def data(n,d):
  X1 = np.random.uniform(0,10,size = n).reshape([-1,1])
  predictors = []
  for i in range(0,d+1):
    predictors.append(X1**i)
  X = np.hstack(predictors)
  Y = np.random.uniform(0,10,size = n).reshape([-1,1])
  return [X,Y]

def curve(w):
  x1 = np.arange(0,10,.01).reshape([-1,1])
  predictors = []
  for i in range(0,len(w)):
    predictors.append(x1**i)
  x = np.hstack(predictors)
  y = np.matmul(x,w)
  return [x,y]

```

```python
X,Y = data(4,3)
plt.scatter(X[:,1],Y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
regr = linear_model.Ridge(alpha = 100)
regr.fit(X,Y)
intercept = regr.intercept_
```

```python
coef = regr.coef_[0][1:]
```

```python
w = np.hstack([intercept,coef])
```

```python
x,y = curve(w)
plt.scatter(X[:,1],Y)
plt.plot(x[:,1],y,color="red")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
