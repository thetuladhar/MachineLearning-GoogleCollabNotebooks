```python
p=np.arange(.01,1,.01)
```

```python
plt.plot(p,p)
plt.show()

plt.plot(p,p/(1-p))
plt.show()

plt.plot(p,np.log(p/(1-p)))
plt.show()
```

Logistics Curve ^

```python
x=np.arange(-10,10,.01)
y=2*x+1

plt.plot(x,y)
plt.show()

plt.plot(x,np.exp(y))
plt.show()

plt.plot(x,np.exp(y)/(1+np.exp(y)))
plt.show()
```

Sigmoid function ^ 

```python
y=2*x+1
plt.plot(x,np.exp(y)/(1+np.exp(y)))
plt.xlim(-10,10)
plt.plot([-10,10],[.5,.5],color="black")
plt.plot([-.5,-.5],[0,1],color="red")
plt.show()
```
