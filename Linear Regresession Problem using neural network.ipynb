```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
```

```python
X=np.random.uniform(0,10,size=100).reshape([-1,1])
Y= 5*X+13 + np.random.normal(0,20,size=100).reshape([-1,1]) #SD OF 20 added random 
```

```python
Xval=np.random.uniform(0,10,size=100).reshape([-1,1])
Yval= 5*X+13 + np.random.normal(0,20,size=100).reshape([-1,1]) #SD OF 20 added random 
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
regr=linear_model.LinearRegression()
```

```python
regr.fit(X,Y)
```

```python
regr.intercept_[0]
```

```python
regr.coef_[0][0]
```

```python
Yhat=regr.predict(X)
```

```python
MSE=mean_squared_error(Y,Yhat)
MSE
```

```python
model = Sequential()
```

```python
model.add(
    Dense(1,activation= None)
)
```

```python
model.compile(
    loss = "mean_squared_error",
    optimizer = RMSprop(learning_rate= .1)
)
```

```python
model.fit(
    X,
    Y,
    epochs = 200,
    validation_data =(Xval,Yval)
)
```
