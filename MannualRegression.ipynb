```python
import numpy as np
import matplotlib.pyplot as plt
```

$
y=w_1x+w_0\\
loss(w_1,w_0)=?\\
\frac{∂loss(w_1,w_0)}{∂w_1}=?\\
\frac{∂loss(w_1,w_0)}{∂w_0}=?\\
$

```python
w=.5
X=np.array([2,5,8,10])
Y=np.array([3,6,2,9])

x=np.arange(0,10,.1)

def loss(w):
  Yhat=w*X
  return np.mean((Y-Yhat)**2)

def dloss_w(w):
  Yhat=w*X
  return np.mean(-2*(Y-Yhat)*X)

#initial plot
plt.scatter(X,Y)
plt.plot(x,w*x,color="red")
plt.ylim(0,12)
plt.show()
print(w,loss(w)) 

def train(lr,epochs):  
  w=0
  for i in range(0,epochs):
    w=w-lr*dloss_w(w)
    plt.scatter(X,Y)
    plt.plot(x,w*x,color="red")
    plt.ylim(0,12)
    plt.show()
    print(w,loss(w))

train(.01,3)
```
