```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn import linear_model 
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay,RocCurveDisplay
```

```python
X,Y=make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1
)
```

```python
color=['blue','red']
color=list(map(lambda y:color[y],Y))
```

```python
Xdf=pd.DataFrame(X)
Ydf=pd.DataFrame(Y)
colordf=pd.DataFrame(color)
df=pd.concat([Xdf,Ydf,colordf],axis=1)
df.columns=["X1","X2","Y","color"]
#df
```

```python
#A way to get the red and blue data frame

red=df[df["color"]=="red"]
blue=df[df["color"]=="blue"]
```

```python
plt.scatter(red["X1"],red["X2"],color="tomato")
plt.scatter(blue["X1"],blue["X2"],color="cornflowerblue")
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.show()
```

```python
regr=linear_model.LogisticRegression()
regr.fit(X,Y)
```

```python
intercept=regr.intercept_
intercept=intercept[0]
intercept
```

```python
slope1=regr.coef_[0][0]
slope2=regr.coef_[0][1]
```

z=slope_1*x_1+ slope_2*x_2 +intercept

intersection is where the equation is equal tp 0 (ie z=0)

```python
#x2=(slope1*x1+intercept)/-slope2
```

```python
x1=np.arange(-3,3,.01)
x2=(slope1*x1+intercept)/(-slope2)

plt.plot(x1,x2,color="purple")

plt.scatter(red["X1"],red["X2"],color="tomato")
plt.scatter(blue["X1"],blue["X2"],color="cornflowerblue")
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.show()
```

```python
ConfusionMatrixDisplay.from_estimator(regr,X,Y)
plt.show()
```

```python
RocCurveDisplay.from_estimator(regr,X,Y)
plt.show()
```
