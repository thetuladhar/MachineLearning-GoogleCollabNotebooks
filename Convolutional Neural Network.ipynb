```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from tensorflow.keras.optimizers import RMSprop
import os
```

```python
!unzip training.zip
```

```python
#TRAINING
X = []
Y = []
for filename in os.listdir("training"):
  if filename[-3:] == "png":
    matrix = np.asarray(Image.open("training/"+filename))/255
    X.append(matrix)
    if filename[:6] == "circle":
      Y.append([1,0])
    else:
      Y.append([0,1])

X = np.array(X)
Y = np.array(Y)
  
```

```python
#TESTING
Xtest = []
Ytest = []
for filename in os.listdir("training"):
  if filename[-3:] == "png":
    matrix = np.asarray(Image.open("training/"+filename))/255
    Xtest.append(matrix)
    if filename[:6] == "circle":
      Ytest.append([1,0])
    else:
      Ytest.append([0,1])

Xtest = np.array(X)
Ytest = np.array(Y)
```

```python
#VALIDATION
Xval = []
Yval = []
for filename in os.listdir("validating"):
  if filename[-3:] == "png":
    row = np.asarray(Image.open("validating/"+filename))/255
    Xval.append(row)
    if filename[:6] == "circle":
      Yval.append([1,0])
    else:
      Yval.append([0,1])

Xval = np.array(Xval)
Yval = np.array(Yval)
  
  
```

```python
model = Sequential()
```

```python
model.add(Conv2D(2,(3,3),activation="relu"))
#2 3x3 filter
#NN is looking for features and is loking for basic 
#features like lines and shapes
#incease number of finer features
#eg facial features

model.add(MaxPooling2D(2,2))
#pooling laye help by making the images smaller. 
#MAX POOLING
```

```python
#BLOCK2
model.add(Conv2D(2,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

```python
#BLOCK3
model.add(Conv2D(2,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

```python
#LAYER4
model.add(Conv2D(2,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
#ADDED MORE LAYERS TO AVOID OVERFITTING
```

```python
#flatten the model
model.add(Flatten())
```

```python
#ADD SOFTMAX FUNCTION
model.add(Dense(2,activation="softmax"))
```

```python
model.compile(
    loss = "categorical_crossentropy",
    optimizer = RMSprop(learning_rate= .001),
    metrics = ["accuracy"]
)
```

```python
model.fit(
    X,
    Y,
    epochs = 1000
    #validation_data =(Xval,Yval)
)
```

```python
#TESTING
Xtest = []
Ytest = []
for filename in os.listdir("testing"):
  if filename[-3:] == "png":
    row = np.asarray(Image.open("testing/"+filename))/255
    Xtest.append(row)
    if filename[:6] == "circle":
      Ytest.append(0)
    else:
      Ytest.append(1)

Xtest = np.array(Xtest)
Ytest = np.array(Ytest)
  
```
