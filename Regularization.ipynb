```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X0=np.ones(2).reshape([-1,1])
X1=np.random.uniform(0,10,size=2).reshape([-1,1])
X=np.hstack([X0,X1])
Y=np.random.uniform(0,10,size=2).reshape([-1,1])
```

```python
plt.scatter(X1,Y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x0=np.ones(1000).reshape([-1,1])
x1=np.arange(0,10,.01).reshape([-1,1])
x=np.hstack([x0,x1])

def loss(w):
  Yhat=np.matmul(X,w)
  return np.mean((Y-Yhat)**2)

def dloss(w):
  Yhat=np.matmul(X,w)
  return -2*np.mean((Y-Yhat)*X,axis=0).reshape([-1,1])

def train(lr,epochs):
  w=np.zeros(2).reshape([-1,1])
  for i in range(0,epochs):
      plt.scatter(X1,Y)
      plt.plot(x1,np.matmul(x,w),color='red')
      plt.xlim(0,10)
      plt.ylim(0,10)
      plt.show()
      w=w-lr*dloss(w)
      print(w,loss(w))

#train(.01,100)
```

```python
regr=linear_model.LinearRegression()
regr.fit(X,Y)
intercept=regr.intercept_[0]
slope=regr.coef_[0][1]
```

```python
x=np.arange(0,10,.01)
y=slope*x+intercept
plt.scatter(X1,Y)
plt.plot(x,y,color='red')
plt.xlim(0,10)
plt.ylim(0,10)
```
