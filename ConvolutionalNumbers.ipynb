```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from tensorflow.keras.optimizers import RMSprop
import os
```

```python
def f(n):
  v=[]
  for i in range(0,10):
    if(i == n):
      v.append(1)
    else:
      v.append(0)
  return v
```

```python
#TRAINING
Xtrain = []
Ytrain = []
for filename in os.listdir("train"):
  if filename[-3:] == "png":
    matrix=np.asarray(Image.open("train/"+filename))/255
    Xtrain.append(matrix)
    n = int(filename[0])
    Ytrain.append(f(n))
  
Xtrain=np.array(Xtrain)
Ytrain=np.array(Ytrain)
```

```python
#VALIDATION
Xval = []
Yval = []
for filename in os.listdir("valid"):
  if filename[-3:] == "png":
    matrix=np.asarray(Image.open("valid/"+filename))/255
    Xval.append(matrix)
    n = int(filename[0])
    Yval.append(f(n))
  
Xval=np.array(Xval)
Yval=np.array(Yval)
```

```python
#TESTING
Xtest = []
Ytest = []
for filename in os.listdir("test"):
  if filename[-3:] == "png":
    matrix=np.asarray(Image.open("test/"+filename))/255
    Xtest.append(matrix)
    n = int(filename[0])
    Ytest.append(n)
  
Xtest=np.array(Xtest)
Ytest=np.array(Ytest)
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32,(3,3),activation="relu"))
model.add(Conv2D(32,(3,3),activation="relu"))
#2 3x3 filter
#NN is looking for features and is looking for basic 
#features like lines and shapes
#incease number of finer features
#eg facial features
model.add(MaxPooling2D(2,2))
#pooling laye help by making the images smaller. 
#MAX POOLING
```

```python
model.add(Conv2D(64,(3,3),activation="relu"))
model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(128,(3,3),activation="relu"))
model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

```python
#flatten the model
model.add(Flatten())
```

```python
#Hidden DENSE LAYER
model.add(Dense(128,activation="relu"))
```

```python
#10 because we have 10 classification
model.add(Dense(10,activation="softmax"))
```

```python
model.compile(
    loss = "categorical_crossentropy",
    optimizer = RMSprop(learning_rate= .001),
    metrics = ["accuracy"]
)
```

```python
model.fit(
    Xtrain,
    Ytrain,
    epochs = 500,
    validation_data =(Xval,Yval)
)
```

```python
m=model.predict(Xtest)
m
```

```python
#Prediction
Yhat=np.argmax(m,axis=1)
Yhat
```
