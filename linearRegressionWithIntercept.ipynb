```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

$
  y = w_1x+w_0 \\
  loss(w_1,w_0)= \frac{1}{n}\sum(Y_i-\hat{Y_i})^2=\frac{1}{n}\sum(Y_i-w_1X_i-w_0)^2 \\
  \frac{\partial loss(w_1,w_0)}{\partial w_1}=\frac{1}{n}\sum-2(Y_i-w_1X_i-w_0)X_i \\
  \frac{\partial loss(w_1,w_0)}{\partial w_0}=\frac{1}{n}\sum-2(Y_i-w_1X_i-w_0)
$

```python
X = np.array([2,5,8,10])
Y = np.array([15,10,13,4])
x = np.arange(0,10,.1)
```

```python
plt.scatter(X,Y)
plt.ylim(0,16)
plt.show()
```

```python
def loss(w1,w0):
  Yhat = w1*X+w0
  return np.mean((Y - Yhat)**2)

def dlossw1(w1,w0):
  Yhat = w1*X+w0
  return np.mean(-2*(Y - Yhat)*X)

def dlossw0(w1,w0):
  Yhat = w1*X+w0
  return np.mean(-2*(Y - Yhat))

def train(lr,epochs):
  w1 = 0
  w0 = 0
  for i in range(0,epochs):
    plt.scatter(X,Y)
    plt.plot(x,w1*x+w0,color="red")
    plt.ylim(0,16)
    plt.show()
    print(w1,w0,loss(w1,w0))
    neww1 = w1 - lr*dlossw1(w1,w0)
    neww0 = w0 - lr*dlossw0(w1,w0)
    w1 = neww1
    w0 = neww0

train(.1,100)
```
