loss(w)=\frac{1}{n}\sum(Y_i-\hat{Y_i})^2
$

$
\frac{partial loss(w){\partial w}=\frac{1}{n}\sum 2(Y_i-wX_i)\\
\sum -2(Y_i-wX_i)X_i\\
$

```python
X=np.array([1.27618731, 1.91988281, 0.29980066, 4.15499504, 3.05893646,
2.88441581, 1.83626965, 4.36963149, 4.80216025, 4.74413713])

Y=np.array([7.96278999, 6.04205697, 9.42108003, 6.06587042, 8.12739562,
6.07481819, 8.07376457, 5.48249222, 6.31457657, 5.34420957])
```

```python
#from logging import logThreads
w=2
#X=np.array([2,5,8,10])
#Y=np.array([3,6,2,9])

x=np.arange(0,10,.1)

def loss(w):
  Yhat=w*X
  return np.mean((Y-Yhat)**2)

def dloss_w(w):
  Yhat=w*X
  return np.mean(-2*(Y-Yhat)*X)

#initial plot
plt.scatter(X,Y)
plt.plot(x,w*x,color="red")
plt.ylim(0,12)
plt.show()
print(w,loss(w)) 


def train(lr,epochs):  
  for i in range(0,epochs):
    w=w-lr*dloss_w(w)
    plt.scatter(X,Y)
    plt.plot(x,w*x,color="red")
    plt.ylim(0,12)
    plt.show()
    print(w,loss(w))
  
```

```python
train(0.1,10)
```
